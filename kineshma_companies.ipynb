```python
# Группировка выручки по отраслям
rev_group = df[['BRANCH', 'Rev2021', 'Rev2020']].groupby(by = 'BRANCH').sum()
rev_group.sort_values(by = 'Rev2021', ascending= False)
```

```python
# Средняя выручка по отраслям
rev_group = df[['BRANCH', 'Rev2021', 'Rev2020']].groupby(by = 'BRANCH').mean()
rev_group.sort_values(by = 'Rev2021', ascending= False)
```

```python
# Количество компаний в каждой отрасли
rev_group = df[['BRANCH', 'Rev2021']].groupby(by = 'BRANCH').count()
rev_group.sort_values(by = 'Rev2021', ascending= False)
```
